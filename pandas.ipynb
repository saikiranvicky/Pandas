#Making changes to the data
Adding a column to the csv file

```
data['Total'] = data['HP']+data['Attack']+data['Defense']
```

```
data.head(3)
```

Drop a coulmn 

```
data = data.drop(columns=['Total'])
data.head(3)
```

Adding all rows of some specific columns with horizontal axis

```
data['Total'] = data.iloc[:,4:10].sum(axis=1)
```

changing columns in the csv file

```
cols = list(data.columns)
print(cols)
data = data[cols[0:4]+[cols[-1]]+cols[4:12]]
data.head(3)
```

#Save the changes to new csv file

#Filtering Data

Conditions to access specific columns

```
new_data = data.loc[(data['Type 1']=='Water') & (data['Type 2']=='Poison')]
```

```
new_data.reset_index(drop=True,inplace=True)
```

```
new_data
```

TO access some elements with matching string

```
data.loc[data['Name'].str.contains('Mega')]
```

To exclude matching strings from the csv file 

```
data.loc[~data['Name'].str.contains('Mega')]
```
